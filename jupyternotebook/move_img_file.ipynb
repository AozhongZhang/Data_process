```python
import os
import shutil

img_file = "H:/BaiduNetdiskDownload/face_SIAT/all_IR_p_faces.txt"
# img_file = "H:/BaiduNetdiskDownload/face_SIAT/all_VIS_faces.txt"

with open(img_file,'r') as f:
    lines = f.readlines()
for line in lines:
    line = line.strip()
    img_id = line.split('/')[4].split('-')[0]
    print(line)
    print(img_id)    
    
    target_path = line.replace("face_SIAT", "face_SIAT/frontal-140")
#     print(int(img_id))
#     target_path = target_path.split('/VIS/')[0] + "/VIS/" + str(int(img_id)) + "/" + target_path.split('/VIS/')[1]    
    target_path = target_path.split('/IR_p/')[0] + "/IR_p/" + str(int(img_id)) + "/" + target_path.split('/IR_p/')[1]

    print(target_path)
    target_dir = os.path.dirname(target_path)
    print(target_dir)
    if not os.path.exists(target_dir):
        os.makedirs(target_dir)
    shutil.copyfile(line, target_path)
```
