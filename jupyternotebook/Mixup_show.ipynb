```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
from scipy.stats import beta
```

```python
im1 = np.array(Image.open("C:/Users/GYGA/Desktop/708-230163.bmp")) 
im2 = np.array(Image.open("C:/Users/GYGA/Desktop/708-136.jpg"))
```

```python
plt.figure(num=1, figsize=(10,5))

plt.subplot(1,2,1)
plt.imshow(im1)
plt.subplot(1,2,2)
plt.imshow(im2)
plt.show()
```

```python
plt.figure(num=2, figsize=(10,10))

for i in range(1,10):
    lam= i*0.1
    im_mixup = (im1*lam+im2*(1-lam)).astype(np.uint8)
    plt.subplot(3,3,i)
    plt.imshow(im_mixup)
plt.show()
```
